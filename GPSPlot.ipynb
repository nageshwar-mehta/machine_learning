```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('GPSData.csv')
df
```

```python
time=df['Time(seconds)']
lat=df['Latitude ']
long=df['Longitude']
lat
# plt.plot(lat)
```

```python
mean_lat = np.mean(df['Latitude '])
mean_lat
```

```python
rel_lat_error = abs((lat - mean_lat)/mean_lat)
rel_lat_error
```

```python
per_error_latitude = np.mean(rel_lat_error)*100
per_error_latitude
```

```python
mean_long = np.mean(df['Longitude'])
mean_long
```

```python
relative_long_error = abs((df['Longitude'] - mean_long)/mean_long)
relative_long_error
```

**taking mean value as True value**

```python
per_error_longitude = np.mean(relative_long_error)*100
per_error_longitude
```

```python
lat
```

```python
plt.figure(figsize=[12,5])
plt.title("GPS DATA (without GPS Dome and without jammer)",fontsize=20,fontfamily="Times New Roman",fontweight="bold")
plt.xlabel("time (in sec)--->",fontsize=10,fontweight='bold')
plt.ylabel("Co-ordinates---->",fontsize=10,fontweight='bold')
plt.ylim(32.8005,32.8015)
plt.plot(time,lat,label='latitude',linewidth = 0.5,color = 'red')
plt.scatter(time,lat,marker="*",color = 'red')
plt.text(20,32.8012,f'Mean percentage Error : {per_error_latitude} %',color = "green",fontweight = 'bold',fontsize=10)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[12,5])
plt.title("GPS DATA (without GPS Dome and without jammer)",fontsize=20,fontfamily="Times New Roman",fontweight="bold")
plt.xlabel("time (in sec)--->",fontsize=10,fontweight='bold')
plt.ylabel("Co-ordinates---->",fontsize=10,fontweight='bold')
plt.plot(time,long,label='longitude',linewidth = 0.5,color = 'blue')
plt.scatter(time,long,marker="+",color = 'blue')
plt.ylim(74.897,74.898)
plt.text(20,74.8976,f'Mean percentage Error : {per_error_longitude} %',color = "green",fontweight = 'bold',fontsize=10)
plt.legend()
plt.show()
```
