```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
df.head()
```

```python
# df
# df.income
```

```python
%matplotlib inline
plt.title("canada's per capita income")
plt.xlabel("year")
plt.ylabel("Income")

plt.scatter(df.year,df.income,color="red",marker="+")
plt.plot(df.year,df.income,color="grey")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
reg.predict([[2020]])
```

```python
c=reg.intercept_
m=reg.coef_
y=m*df.year+c
```

```python
plt.figure(figsize=[12,3])
plt.title("Canada's Per Capita Income")
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(df.year, df.income, color="red", marker="+",label="actual Income")
plt.plot(df.year, y, color="green",label="Linear Regession line ")
plt.legend()
plt.show()



```
